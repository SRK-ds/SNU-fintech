```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
pd.read_csv('(130)rf_metric_인천공항(17687) (1).csv')
```

```python
df = pd.read_csv('(130)rf_predict_인천공항(17687) (1).csv')
```

```python
# ITEM_CD별로 그래프 그리기
items = df['ITEM_CD'].unique()

for item in items:
    item_df = df[df['ITEM_CD'] == item]

    plt.figure(figsize=(10, 6))

    # SALE_QTY와 predict 같은 그래프에 플롯
    plt.plot(item_df['YMD_CD'], item_df['SALE_QTY'], 'b-', label='SALE_QTY')
    plt.plot(item_df['YMD_CD'], item_df['predict'], 'r--', label='predict')

    # 레이블과 타이틀 추가
    plt.xlabel('Date')
    plt.ylabel('Values')
    plt.title(f'ITEM_CD: {item}')
    plt.legend()

    # 그래프 보여주기
    plt.show()
```
